# 1. Products of what prce range are popular in which regions?

```python
# Group by region and price range
merged_df['PriceRange'] = pd.cut(merged_df['Price_y'], bins=[0, 100, 500, 1000, 5000], 
                                 labels=['Low (0-100)', 'Medium (100-500)', 'High (500-1000)', 'Very High (>1000)'])
price_region_popularity = merged_df.groupby(['Region', 'PriceRange']).size().reset_index(name='Count')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=price_region_popularity, x='Region', y='Count', hue='PriceRange')
plt.title('Product Popularity by Price Range and Region')
plt.ylabel('Number of Transactions')
plt.xlabel('Region')
plt.legend(title='Price Range')
plt.show()
```

# 2. In which month do transactions go up or down?

```python
# Extract month from TransactionDate
merged_df['TransactionMonth'] = pd.to_datetime(merged_df['TransactionDate']).dt.month

# Group by month
monthly_transactions = merged_df['TransactionMonth'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_transactions.index, y=monthly_transactions.values)
plt.title('Monthly Transaction Trends')
plt.ylabel('Number of Transactions')
plt.xlabel('Month')
plt.xticks(range(1, 13))
plt.show()
```

# 3. Price of prdoucts vs popularity

```python
# Group by product and calculate popularity
price_popularity = merged_df.groupby('ProductName').agg(
    Popularity=('TransactionID', 'count'), AvgPrice=('Price_y', 'mean')).reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=price_popularity, x='AvgPrice', y='Popularity')
plt.title('Product Price vs Popularity')
plt.xlabel('Average Price (USD)')
plt.ylabel('Popularity (Transaction Count)')
plt.show()
```

# 4. Avg Price of things in a category vs popularity

```python
# Group by category and calculate metrics
category_popularity = merged_df.groupby('Category').agg(
    AvgPrice=('Price_y', 'mean'), Popularity=('TransactionID', 'count')).reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=category_popularity, x='Category', y='AvgPrice', hue='Popularity')
plt.title('Avg Price vs Popularity by Category')
plt.ylabel('Average Price (USD)')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

# 5. In which months do signups go up or down


```python
# Extract month from SignupDate
customers_df['SignupMonth'] = pd.to_datetime(customers_df['SignupDate']).dt.month

# Count signups per month
monthly_signups = customers_df['SignupMonth'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_signups.index, y=monthly_signups.values)
plt.title('Monthly Signup Trends')
plt.ylabel('Number of Signups')
plt.xlabel('Month')
plt.xticks(range(1, 13))
plt.show()
```

# 6. Smartwatch is the most bought product, so where is it least bought, which month is it least bought in 


```python
# Filter for smartwatch transactions
smartwatch_df = merged_df[merged_df['ProductName'].str.contains('Smartwatch', case=False, na=False)]

# Group by region and month
region_month_sales = smartwatch_df.groupby(['Region', 'TransactionMonth']).size().reset_index(name='Count')

# Least bought region and month
least_bought_region = region_month_sales.loc[region_month_sales['Count'].idxmin()]

print(f"Least bought region: {least_bought_region['Region']}, Month: {least_bought_region['TransactionMonth']}")
```

# 7. Are there any regions where specific products are significantly underperforming?


```python
# Group by product and region
product_region_sales = merged_df.groupby(['ProductName', 'Region']).size().reset_index(name='Count')

# Identify underperforming products (low quartile)
threshold = product_region_sales['Count'].quantile(0.25)
underperforming = product_region_sales[product_region_sales['Count'] <= threshold]

print("Underperforming products by region:")
print(underperforming)
```

# 8. Which products are most frequently purchased together?


```python
# Analyze co-occurrences of products in transactions
from itertools import combinations
from collections import Counter

# Group products by transaction
transaction_products = merged_df.groupby('TransactionID')['ProductName'].apply(list)

# Find co-occurrences
pair_counts = Counter()
for products in transaction_products:
    for pair in combinations(products, 2):
        pair_counts[pair] += 1

# Most frequent pairs
most_frequent_pairs = pair_counts.most_common(10)
print("Most frequently purchased product pairs:", most_frequent_pairs)
```

# 9. Which products contribute the most to total revenue?


```python
# Group by product and calculate total revenue
product_revenue = merged_df.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 6))
product_revenue.head(10).plot(kind='bar')
plt.title('Top 10 Products by Revenue')
plt.ylabel('Total Revenue (USD)')
plt.xlabel('Product Name')
plt.xticks(rotation=45)
plt.show()
```

# 10. What percentage of transactions comes from repeat customers?


```python
# Identify repeat customers
repeat_customers = merged_df.groupby('CustomerID').size()
repeat_percentage = (repeat_customers[repeat_customers > 1].count() / repeat_customers.count()) * 100

print(f"Percentage of transactions from repeat customers: {repeat_percentage:.2f}%")
```

# 11. Which regions have the highest/lowest signup-to-transaction conversion rates?


```python
# Calculate conversion rates
region_signups = customers_df['Region'].value_counts()
region_transactions = merged_df['Region'].value_counts()

conversion_rates = (region_transactions / region_signups * 100).dropna()
highest_conversion = conversion_rates.idxmax()
lowest_conversion = conversion_rates.idxmin()

print(f"Highest conversion rate region: {highest_conversion}")
print(f"Lowest conversion rate region: {lowest_conversion}")
```

# 12. Is there a correlation between product price and transaction quantity?


```python
# Correlation analysis
correlation = merged_df[['Price_y', 'Quantity']].corr().iloc[0, 1]
print(f"Correlation between product price and transaction quantity: {correlation:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=merged_df, x='Price_y', y='Quantity', scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})

# Add title and labels
plt.title('Correlation between Product Price and Transaction Quantity', fontsize=14)
plt.xlabel('Product Price (USD)', fontsize=12)
plt.ylabel('Transaction Quantity', fontsize=12)

# Annotate with correlation value
plt.text(0.8 * max(merged_df['Price_y']), 0.8 * max(merged_df['Quantity']), 
         f"Correlation: {correlation:.2f}", fontsize=12, color='darkred', bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

# 13. Which products are frequently purchased in bulk (high quantities in a single transaction)?

```python
# Filter for bulk purchases (e.g., Quantity > 10)
bulk_purchases = merged_df[merged_df['Quantity'] > 10]

if bulk_purchases.empty:
    print("No bulk purchases (Quantity > 10) found in the dataset.")
else:
    # Get the count of bulk purchases for each product
    bulk_products = bulk_purchases['ProductName'].value_counts()

    # Plot the top 10 products frequently purchased in bulk
    plt.figure(figsize=(12, 6))
    bulk_products.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Top 10 Products Frequently Purchased in Bulk', fontsize=14)
    plt.ylabel('Number of Bulk Purchases', fontsize=12)
    plt.xlabel('Product Name', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.show()
```
